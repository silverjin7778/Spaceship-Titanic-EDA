```python
import pandas as pd
import numpy as np
import os
```

```python
DATA_DIR = 'dataset'
datalist = []
for filename in os.listdir('dataset')[::-1]:
    if filename in ['train.csv', 'test.csv']:
        filepath = os.path.join(DATA_DIR, filename)
        df = pd.read_csv(filepath)
        datalist.append(df)
    if filename in ['train.csv']:
        filepath = os.path.join(DATA_DIR, filename)
        train = pd.read_csv(filepath)
    if filename in ['test.csv']:
        filepath = os.path.join(DATA_DIR, filename)
        test = pd.read_csv(filepath)


merge_data = pd.concat(datalist)
y_value = merge_data.pop('SalePrice').values
y_value = y_value[:1460]
merge_data = merge_data.drop(columns='Id', axis =1)
merge_data = merge_data.reset_index(drop=True)
```

```python
#제거할 열은 제거하기
total_data_counts = merge_data.shape[0]
cond = (merge_data.isna().sum()>0).values
    #결측값 비율이 60%이상인 열은 삭제
delete_list = (merge_data.isna().sum() / total_data_counts)[cond].sort_values(ascending = False)[:6].index
merge_data = merge_data.drop(columns=delete_list, axis = 1)
```

```python
merge_data['LotFrontage'].fillna(merge_data['LotFrontage'].mean(), inplace=True)
merge_data['BsmtExposure'].fillna(merge_data['BsmtExposure'].mode()[0], inplace = True)
```

```python
isna_col_names = merge_data.isna().sum().sort_values(ascending=False).index
for isna_col_name in isna_col_names:
    if merge_data[isna_col_name].dtype == 'float64':
        merge_data[isna_col_name].fillna(merge_data[isna_col_name].mean(), inplace=True)
    elif merge_data[isna_col_name].dtype == 'object':
        merge_data[isna_col_name].fillna(merge_data[isna_col_name].mode()[0], inplace=True)
```

```python
merge_data_encoding = merge_data.copy()
drop_col_list = []
for col_name in merge_data.columns:
    if merge_data[col_name].dtype == 'object':
        drop_col_list.append(col_name)
        dummy_df = pd.get_dummies(merge_data[col_name])
        merge_data_encoding = pd.concat([merge_data_encoding, dummy_df], axis = 1)

merge_data_encoding
```
